```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'sales_data.csv',encoding='ISO-8859-1')
df
```

```python
df.columns
```

```python
df.columns=['User_ID', 'Cust_name', 'Product_ID', 'Gender', 'Age Group' , 'Age','Marital', 'State', 'Zone', 'Occupation', 'Product_Category','Orders', 'Amount', 'Status', 'unnamed1']
df
```

```python
df[df['Amount'].isnull()]
```

```python
df = df.dropna(subset=['Amount'])
df
```

```python
print(df['Amount'].isnull().sum())
```

```python
df=df.drop(['Status','unnamed1'],axis=1)
df
```

```python
df.info()
```

```python
new_values ={0 :'single',1: 'married'}
```

```python
df['Marital']= df['Marital'].map(new_values)
```

```python
df
```

```python
df['Gender'].unique()
```

```python
gen ={'M' :'Male','F': 'Female'}
```

```python
df['Gender']= df['Gender'].map(gen)
```

```python
df
```

```python
plt.figure(figsize=(6, 4))
a=sns.countplot(x='Gender', data=df,hue='Marital' ,palette='pastel')
plt.title('Count of Males and Females')
ax=plt.gca()
ax.set_facecolor('lavender')
a.figure.set_facecolor('grey')
plt.show()
```

```python
grouped_data = df.groupby(['Gender', 'Age Group', 'State', 'Zone','Occupation', 'Product_Category', 'Marital']).size().reset_index(name='count')
print(grouped_data.head())
```

```python
plt.figure(figsize=(18, 14), facecolor='lavender')

# Use the original df and group by relevant fields for each chart
# 1. Gender
plt.subplot(3, 3, 1)
axes = plt.gca()
axes.set_facecolor('red')
gender_amt = df.groupby(['Gender', 'Marital'])['Amount'].sum().reset_index()
sns.barplot(data=gender_amt, x='Gender', y='Amount', hue='Marital', palette='pastel')
plt.title("Gender vs Total Amount (by Marital Status)")

# 2. Age Group vs Gender
plt.subplot(3, 3, 2)
axes = plt.gca()
axes.set_facecolor('hotpink')
age_gender_amt = df.groupby(['Age Group', 'Gender'])['Amount'].sum().reset_index()
sns.barplot(data=age_gender_amt, x='Age Group', y='Amount', hue='Gender', palette='coolwarm')
plt.title("Age Group vs Total Amount (by Gender)")

# 3. Top 10 States vs Gender
plt.subplot(3, 3, 3)
axes = plt.gca()
axes.set_facecolor('grey')
state_amt = df.groupby(['State', 'Gender'])['Amount'].sum().reset_index()
top_states = state_amt.groupby('State')['Amount'].sum().nlargest(10).index
filtered_states = state_amt[state_amt['State'].isin(top_states)]
sns.barplot(data=filtered_states, y='State', x='Amount', hue='Gender', palette='Blues_d')
plt.title("Top 10 States by Total Amount (by Gender)")

# 4. Zone vs Gender
plt.subplot(3, 3, 4)
axes = plt.gca()
axes.set_facecolor('lavender')
zone_amt = df.groupby(['Zone', 'Gender'])['Amount'].sum().reset_index()
sns.barplot(data=zone_amt, x='Zone', y='Amount', hue='Gender', palette='Purples')
plt.title("Zone vs Total Amount (by Gender)")

# 5. Occupation vs Gender
plt.subplot(3, 3, 5)
axes = plt.gca()
axes.set_facecolor('navy')
occu_amt = df.groupby(['Occupation', 'Gender'])['Amount'].sum().reset_index()
top_occu = occu_amt.groupby('Occupation')['Amount'].sum().nlargest(10).index
filtered_occu = occu_amt[occu_amt['Occupation'].isin(top_occu)]
sns.barplot(data=filtered_occu, y='Occupation', x='Amount', hue='Gender', palette='magma')
plt.title("Top Occupations by Total Amount (by Gender)")

# 6. Product Category vs Gender
plt.subplot(3, 3, 6)
axes = plt.gca()
axes.set_facecolor('yellow')
cat_amt = df.groupby(['Product_Category', 'Gender'])['Amount'].sum().reset_index()
top_cat = cat_amt.groupby('Product_Category')['Amount'].sum().nlargest(10).index
filtered_cat = cat_amt[cat_amt['Product_Category'].isin(top_cat)]
sns.barplot(data=filtered_cat, y='Product_Category', x='Amount', hue='Gender', palette='cubehelix')
plt.title("Top Product Categories by Total Amount (by Gender)")

# 7. Marital Status vs Gender
plt.subplot(3, 3, 7)
axes = plt.gca()
axes.set_facecolor('orange')
marital_amt = df.groupby(['Marital', 'Gender'])['Amount'].sum().reset_index()
sns.barplot(data=marital_amt, x='Marital', y='Amount', hue='Gender', palette='Set2')
plt.title("Marital Status vs Total Amount (by Gender)")

plt.tight_layout()
plt.show()
```

Occupation and Product Choice: Occupational backgrounds influence purchasing patterns, suggesting a correlation between profession and product   preferences. Tailoring product recommendations based on occupation can enhance customer satisfaction.  

Marital Status: Marital status appears to have an impact on buying behavior. Understanding the preferences of single and married customers   enables personalized marketing efforts.  

Top Categories and Products: Certain product categories and individual items consistently generate high sales volumes. Identifying these top   performers helps optimize inventory management and promotional activities.  

Marketing Implications: These insights inform targeted marketing strategies, customer segmentation, and product development initiatives.   Leveraging this knowledge can lead to increased sales, customer engagement, and business growth.
